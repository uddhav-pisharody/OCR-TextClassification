```python
def take_photo():
    cap = cv2.VideoCapture(0)
    ret, frame = cap.read()
    cv2.imwrite('mobilephoto.jpg', frame)
    cap.release()
```

```python
take_photo
```

```python
cap = cv2.VideoCapture(0) # set up capture device
cap.set(cv2.CAP_PROP_FRAME_WIDTH, 640) # set capture device properties
cap.set(cv2.CAP_PROP_FRAME_HEIGHT, 480)

while True:
    ret, frame = cap.read() # capture a frame from the camera
    cv2.imshow('frame', frame) # display the captured frame
    if cv2.waitKey(1) == ord('q'): # exit loop if 'q' is pressed
        break

cap.release() # release the capture device
cv2.destroyAllWindows() # close all windows
```
